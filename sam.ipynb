```python
import cv2
import time
```

```python

face_capture = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

video_capture = cv2.VideoCapture(0)

video_capture.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
video_capture.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

prev_time = 0

while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_capture.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(60, 60)
    )

    curr_time = time.time()
    fps = 1 / (curr_time - prev_time) if prev_time else 0
    prev_time = curr_time

    for (x, y, w, h) in faces:
        # Shadow effect rectangle (for better visual depth)
        cv2.rectangle(frame, (x - 3, y - 3), (x + w + 3, y + h + 3), (0, 0, 0), 4)

        # Main green rectangle with rounded corners
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 80), 2)

        # Add text above face box
        cv2.putText(frame, "Face", (x, y - 10),
                    cv2.FONT_HERSHEY_DUPLEX, 0.7, (255, 255, 255), 2)

    # Overlay FPS counter
    cv2.putText(frame, f"FPS: {int(fps)}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (50, 255, 255), 2)

    # Overlay exit instruction
    cv2.putText(frame, "Press 'A' to Exit", (10, frame.shape[0] - 10),
                cv2.FONT_HERSHEY_DUPLEX, 0.7, (200, 200, 255), 1)

    # Display output
    cv2.imshow("Human Face Detection", frame)

    if cv2.waitKey(10) & 0xFF == ord('a'):
        break

# Release resources
video_capture.release()
cv2.destroyAllWindows()
```
